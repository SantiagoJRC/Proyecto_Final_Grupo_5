```python
import pandas as pd
```

```python
df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
df.head(23)
```

```python
def documento(n):
    n1 = str(n[0])
    s = 0
    for i in n1:
        s += int(i)
    return s
```

```python
df['Promedio ND'] = df.apply(documento, axis =1)
df.head(23)
```

```python
def Punto1(x):
    a = x * 105**2 +2*105 + 23
    b = x * 100**2 +2*100 + 23
    return (a - b)/(5)
```

```python
df['Punto 1'] = df['Promedio ND'].apply(Punto1)
df.head(23)
```

```python
def califiación1(n):
    n1 = str(n[1])
    n2 = str(n[5])
    if n2 == n1:
        return('5')
    else:
        return('0')
```

```python
df['Calificación'] = df.apply(califiación1, axis = 1)
df.head(23)
```

```python
df.to_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv", index=False)
```
