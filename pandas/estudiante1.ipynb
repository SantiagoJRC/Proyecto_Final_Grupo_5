```python
import pandas as pd
```

```python
df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
df.head(23)
```

```python
def p2(x):
    a = x * 101**2 +2*101 + 23
    b = x * 100**2 +2*100 + 23
    return (a - b)/(101 - 100)
```

```python
df['Punto 2'] = df['Promedio ND'].apply(p2)
df.head(18)
```

```python
def notas2(n):
    n1 = str(n[2])
    n2 = str(n[7])
    if n2 == n1:
        return('5')
    else:
        return('0')
```

```python
df['CP2'] = df.apply(notas2, axis = 1)
df.head(18)
```

```python
df.to_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv", index=False)
```

```python
col_idx = 3

df.iloc[:, col_idx] = (
    df.iloc[:, col_idx]
      .astype(str)
      .str.replace(",", ".")
      .astype(float)
)

def p3(fila):
    return 2 * fila["Promedio ND"] * 100 + 2

df["Punto 3"] = df.apply(p3, axis=1)

df.head(18)
```

```python
def CP3(n):
    n1 = str(n[9])
    n2 = str(n[3])
    return '5' if n1 == n2 else '0'

df["CP3"] = df.apply(CP3, axis=1)
df.head(23)
```
