```python
import pandas as pd
```

```python
df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
df.head(23)
```

```python
def p2(x):
    a = x * 101**2 +2*101 + 23
    b = x * 100**2 +2*100 + 23
    return (a - b)/(101 - 100)
```

```python
df['Punto 2'] = df['Promedio ND'].apply(p2)
df.head(18)
```

```python
def notas2(n):
    n1 = str(n[2])
    n2 = str(n[7])
    if n2 == n1:
        return('5')
    else:
        return('0')
```

```python
df['CP2'] = df.apply(notas2, axis = 1)
df.head(18)
```

```python
df["Promedio ND"] = df["Promedio ND"].astype(str).str.replace(",", ".").astype(float)
```

```python
def cm(x):
    c=(2*100*x) + 2
    return float(c)

df["Punto 3"]=df["Promedio ND"].apply(cm)
df
```

```python
def Notas3(x):
     if x["Punto 3"]==float(x["Sea la función de costo C(x) = n*x^2 + 2*x + 23.\xa0\nUsar el costo marginal para aproximar el costo promedio entre x= 100 y x =101 (Poner dos decimales)"]):
        return 5
     else:
        return 0
df["CP3"]=df.apply(Notas3,axis=1)
df
```

```python
df.to_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv", index=False)
```
